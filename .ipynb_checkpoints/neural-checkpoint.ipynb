```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from nltk.stem import PorterStemmer
from nltk import word_tokenize
```

```python
import string
```

```python
def process_text(text, stem=True):
    """ Tokenize text and stem words removing punctuation """
    text = text.translate(string.punctuation)
    tokens = word_tokenize(text)
 
    if stem:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(t) for t in tokens]
 
    return tokens
```

```python
df = pd.read_csv('NYTimesBlogTrain.csv', index_col=9)
```

```python
df['PubDate'] = pd.to_datetime(df['PubDate'])
```

```python
df['day'] = df['PubDate'].map(lambda x: x.weekday())
```

```python
df['saturday'] = (df['day'] == 5).astype(int)
df['sunday'] = (df['day'] == 6).astype(int)
df['friday'] = (df['day'] == 4).astype(int)
df['thursday'] = (df['day'] == 3).astype(int)
df['wednesday'] = (df['day'] == 2).astype(int)
df['tuesday'] = (df['day'] == 1).astype(int)
df['monday'] = (df['day'] == 0).astype(int)
```

```python
df.groupby(df['PubDate'].map(lambda x: x.weekday())).agg('mean')
```

```python
df['c1'] = ((df['NewsDesk'] == 'Business') & (df['SectionName'] == 'Business Day') & (df['SubsectionName'] == 'Dealbook')).astype(int)
df['c2'] = ((df['NewsDesk'] == 'Business') & (df['SectionName'] == 'Business Day') & (df['SubsectionName'] == 'Small Business')).astype(int)
df['c3'] = ((df['NewsDesk'] == 'Business') & (df['SectionName'] == 'Crosswords/Games')).astype(int)
df['c4'] = ((df['NewsDesk'] == 'Business') & (df['SectionName'] == 'Technology')).astype(int)
df['c5'] = ((df['NewsDesk'] == 'Styles') & (df['SectionName'] == 'Health')).astype(int)
df['c6'] = ((df['NewsDesk'] == 'Styles') & (df['SectionName'] == 'U.S.')).astype(int)
df['c7'] = (df['NewsDesk'] == 'TStyle').astype(int)
df['c8'] = (df['NewsDesk'] == 'Culture').astype(int)
df['c9'] = (df['NewsDesk'] == 'OpEd').astype(int)
df['c10'] = (df['NewsDesk'] == 'Foreign').astype(int)
df['c11'] = (df['NewsDesk'] == 'Metro').astype(int)
df['c12'] = (df['NewsDesk'] == 'Science').astype(int)
df['c13'] = (df['NewsDesk'] == 'Travel').astype(int)
df['unknown'] = (df['NewsDesk'].isnull()).astype(int)
df['other'] = 1 - (df['c1'] + df['c2'] + df['c3'] + df['c4'] + df['c5'] + df['c6'] + df['c7'] + df['c8'] + df['c9'] + df['c10'] + df['c11'] + df['c12'] + df['c13'] + df['unknown'])
```

```python
pd.set_option('max_columns', 50)
```

```python
features = ['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10', 'c11', 'c12', 'c13', 'other', 'unknown', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday', 'WordCount']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df.ix[df['Snippet'].isnull(), 'Snippet'] = ''
```

```python
l = df['Snippet'].str.lower() + ' ' + df['Headline'].str.lower()
```

```python
vectorizer = TfidfVectorizer(max_df=0.1, min_df=0.005, stop_words='english', use_idf=True, tokenizer = process_text)
X = vectorizer.fit_transform(l.values)
X = X.toarray()
```

```python
X = np.concatenate((X, df[features].values), axis=1)
```

```python
y = df['Popular'].values
```

```python
from sklearn.cross_validation import StratifiedKFold
```

```python
skf = StratifiedKFold(y, n_folds = 5, shuffle=True)
```

```python
from sklearn.metrics import auc, roc_curve
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
for train_index, test_index in skf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    mod = RandomForestClassifier(n_estimators=50, max_depth = 10, max_features = None).fit(X_train.toarray(), y_train)
    pred = mod.predict_proba(X_test.toarray())
    fpr, tpr, thresholds = roc_curve(y_test, pred[:,0], pos_label=0)
    print auc(fpr, tpr)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
best_auc = 0
best_mod = None
```

```python
for train_index, test_index in skf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    mod = AdaBoostClassifier(n_estimators=150, learning_rate = 0.1).fit(X_train, y_train)
    pred = mod.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, pred[:,0], pos_label=0)
    test_auc = auc(fpr, tpr)
    
    trainpred = mod.predict_proba(X_train)
    t_fpr, t_tpr, t_thresholds = roc_curve(y_train, trainpred[:,0], pos_label=0)
    print 'test', test_auc, 'train', auc(t_fpr, t_tpr)
    
    if test_auc > best_auc:
        best_mod = mod
        best_auc = test_auc
```

```python
best_ada_mod = best_mod
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
for train_index, test_index in skf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    mod = GradientBoostingClassifier(n_estimators=100, learning_rate = 0.1, max_depth=4).fit(X_train, y_train)
    pred = mod.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, pred[:,0], pos_label=0)
    test_auc = auc(fpr, tpr)
    
    trainpred = mod.predict_proba(X_train)
    t_fpr, t_tpr, t_thresholds = roc_curve(y_train, trainpred[:,0], pos_label=0)
    print 'test', test_auc, 'train', auc(t_fpr, t_tpr)
    
    if test_auc > best_auc:
        best_mod = mod
        best_auc = test_auc
```

```python
best_gradboost_mod = best_mod
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
for train_index, test_index in skf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    mod = BaggingClassifier(n_estimators = 20, max_samples = 0.1, oob_score=True).fit(X_train, y_train)
    pred = mod.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, pred[:,0], pos_label=0)
    test_auc = auc(fpr, tpr)
    
    trainpred = mod.predict_proba(X_train)
    t_fpr, t_tpr, t_thresholds = roc_curve(y_train, trainpred[:,0], pos_label=0)
    print 'test', test_auc, 'train', auc(t_fpr, t_tpr)
    
    if test_auc > best_auc:
        best_mod = mod
        best_auc = test_auc
```
